# Number of gold medals awarded to USA for both men and women for all olympic competitions

```python
USA_gold_history = olympic_data[(olympic_data.NOC == 'USA')&(olympic_data.Medal=='Gold')]
USA_gold_history.groupby(['Edition','Gender']).size().unstack('Gender',fill_value=0).plot(kind='bar');
```

# Top 5 athletes with most gold medals won in olympic history

```python
medal_colors = ['#dbb40c', '#c5c9c7', '#a87900']
my_medal_colors = ListedColormap(sns.color_palette(medal_colors))
top5_gold_winners = olympic_data.groupby(['Athlete','Medal']).size().unstack('Medal',fill_value=0)
top5_gold_winners = top5_gold_winners.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']].head(5)
top5_gold_winners.plot(kind='bar',colormap=my_medal_colors)
```

# Medals won per country in 2008 olympics heatmap

```python
all_medals_2008 = olympic_data[olympic_data.Edition == 2008]
all_medals_2008 = all_medals_2008.groupby(['NOC','Medal']).size().unstack('Medal',fill_value=0).sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']]
all_medals_2008 = all_medals_2008.transpose()
all_medals_2008
sns.heatmap(all_medals_2008,cmap='Blues')
```
