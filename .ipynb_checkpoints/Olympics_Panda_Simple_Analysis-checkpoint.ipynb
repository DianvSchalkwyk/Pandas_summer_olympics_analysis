# Import Pandas

```python
import pandas as pd
```

# Read csv data file into od (olympic data)

```python
od = pd.read_csv('C:/Users/dianv/OneDrive/Documents/GitHub/Summer_Olympics_Panda_Analysis/data/olympics.csv')
```

# Display fields to see if data loaded correctly

```python
od.head()
```

# See how many medals were handed out during each olympic competition (1896-2008)

```python
od.Edition.value_counts(ascending=True)
```

# Total medal count for men and women during olympics

```python
od.Gender.value_counts()
```

# Sort names alphabetically for each olympic competition

```python
od.sort_values(by=['Edition','Athlete'])
```

# Total medal type count during olympics

```python
od.Medal.value_counts()
```

# Show all male athletes who won gold medals

```python
od[(od.Medal == 'Gold') & (od.Gender == 'Men')]
```

# Show all female athletes who won gold medals

```python
od[(od.Medal == 'Gold') & (od.Gender == 'Women')]
```

# Total medal count for all the sports

```python
od.Sport.value_counts().sort_index()
```

# Total medal count for all the disciplines

```python
od.Discipline.value_counts().sort_index()
```

# Total medal count for each event of a discipline

```python
od[['Discipline','Event']].value_counts().sort_index()
```

# Medal count per country

```python
od[['NOC','Medal']].value_counts().sort_index()
```

# Country that won the most amount of medals combined

```python
od.NOC.value_counts().nlargest(1)
```

# Athlete medal count, from most to least

```python
od[['Athlete','Medal']].value_counts()
```

# All the events Michael Phelps won a medal in

```python
michael_phelps = od[od.Athlete == 'PHELPS, Michael']
michael_phelps.Event.value_counts()
```

# Show all the athletes who won a gold in the 100m

```python
od[(od.Gender=='Men') & (od.Event=='100m') & (od.Medal=='Gold')][['Edition','City','Athlete','NOC']]
```
